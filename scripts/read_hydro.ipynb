```python
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_hydrobasins(shp_path, color_field=None, figsize=(10, 8)):
    print("Loading shapefile:", shp_path)
    gdf = gpd.read_file(shp_path)

    plt.figure(figsize=figsize)

    if color_field and color_field in gdf.columns:
        gdf.plot(column=color_field, cmap="viridis", legend=True)
        plt.title(f"HydroBASINS colored by {color_field}")
    else:
        gdf.plot(edgecolor='black', facecolor='lightblue')
        plt.title("HydroBASINS basin map")

    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True)
    plt.show()

plot_hydrobasins("E:\\DataSciencePRJ\\Asia\\hybas_as_lev06_v1c.shp", color_field="SUB_AREA")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_hydrobasins(shp_path, color_field=None, figsize=(10, 8)):
    """
    Plots any HydroBASINS shapefile.
    - shp_path: ng dn file .shp
    - color_field: tn ct  t mu (vd: 'SUB_AREA', 'PFAF_ID'), hoc None  t 1 mu
    """
    print("Loading shapefile:", shp_path)
    gdf = gpd.read_file(shp_path)

    plt.figure(figsize=figsize)

    if color_field and color_field in gdf.columns:
        gdf.plot(column=color_field, cmap="viridis", legend=True)
        plt.title(f"HydroBASINS colored by {color_field}")
    else:
        gdf.plot(edgecolor='black', facecolor='lightblue')
        plt.title("HydroBASINS basin map")

    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True)
    plt.show()

# Example usage:
plot_hydrobasins("E:\\DataSciencePRJ\\Asia\\hybas_as_lev09_v1c.shp", color_field="SUB_AREA")
```
